```
# Importando bibliotecas do matplotlib (modulo de graficos)
from matplotlib import pyplot as plt
from matplotlib import ticker
```

```
# Criando um grafico de barras
axis = df_dados.plot(x='Unidade da Federação', y='2008/Ago', kind='bar', figsize=(9,4)) # Criando uma variavel para configurar eixos do grafico
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# Criando titulo do gráfico
plt.title('Valor por Unidade da Federação')

# Mostrando o gráfico
plt.show()
```

### DESAFIO 2

```
# Desafio 1: Executar o gráfico com o mês mais recente

# Desafio 2: Deixar o eixo x do gráfico deitado (45 ou 90 graus)
```

```
df_ultima_data_grafico = df_dados_desafio.iloc[:,[0,-2]]

df_ultima_data_grafico.iloc[:,1] = df_ultima_data_grafico.iloc[:,1].replace({'-': 0})
df_ultima_data_grafico.iloc[:,1] = pd.to_numeric(df_ultima_data_grafico.iloc[:,1], errors='coerce')
```

```
# Criando um grafico de barras
axis_desafio = df_ultima_data_grafico.plot(
    x=df_ultima_data_grafico.columns[0],
    y=df_ultima_data_grafico.columns[1],
    kind='bar',
    figsize=(9,4)) # Criando uma variavel para configurar eixos do grafico

axis_desafio.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# Criando titulo do gráfico
plt.title('Valor Ultimo Mês por Unidade da Federação')
plt.xticks(rotation=45, ha='right', fontsize=10)

# Mostrando o gráfico
plt.show()
```
